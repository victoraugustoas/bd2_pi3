```python
from pyspark.ml.feature import MinHashLSH
from pyspark.ml.linalg import Vectors
from pyspark.sql.functions import col
from pyspark import SparkConf, SparkContext, StorageLevel
from pyspark.sql.session import SparkSession

from sklearn.datasets import fetch_20newsgroups
import csv
import string

import numpy as np

from pprint import pprint
```

```python
def removeChar(line):
    values = string.punctuation
    
    for item in line:
        if item == ' ':
            continue
        if item in values:
            line = line.replace(item, "").encode('utf-8')
            
    values = ['\n','\t','\r']
    for value in values:
        line = line.replace(value, " ").encode('utf-8')
            
    return line
```

```python
config_cluster = SparkConf().setMaster("local").setAppName("bd_pi3")
sc = SparkContext(conf=config_cluster)
spark = SparkSession(sc)
```

### Gerando csv de treino

```python
# subset de treino do sklearn
newsgroups_train = fetch_20newsgroups(subset='train')
```

```python
train_set = [[newsgroups_train.data[i], newsgroups_train.target_names[newsgroups_train.target[i]]] for i in range(len(newsgroups_train.data))]
```

```python
for idx, train in enumerate(train_set):
    train_set[idx][0] = removeChar(train[0])
```

```python
with open('train.csv', 'w') as arq:
    writer = csv.writer(arq, delimiter=';')
    writer.writerow(['text', 'class'])
    for row in train_set:
        writer.writerow(row)
```
